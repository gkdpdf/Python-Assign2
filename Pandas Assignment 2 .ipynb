```python
#10
import pandas as pd
import random
from datetime import timedelta, date

# Define the start date and end date
start_date = date(2023, 9, 20)
end_date = date(2023, 9, 29)

# Generate a list of dates
date_range = [start_date + timedelta(days=x) for x in range(10)]

# Generate random sales values for 10 days
sales = [random.randint(800, 2000) for _ in range(10)]

# Create the DataFrame
data = {'Date': date_range, 'Sales': sales}
df = pd.DataFrame(data)

# Display the DataFrame
print(df)
```

```python
df
```

```python
df = df.sort_values(by='Date')


df['Moving_Avg'] = df['Sales'].rolling(window=7, min_periods=7).mean()

# Display the DataFrame
print(df)
```

```python
import pandas as pd

def add_weekday_column(df):
    # Convert 'Date' column to datetime format (if not already)
    df['Date'] = pd.to_datetime(df['Date'])
    
    # Define a function to get the weekday name
    def get_weekday_name(date):
        return date.strftime('%A')

    # Apply the function to the 'Date' column
    df['Weekday'] = df['Date'].apply(get_weekday_name)
    return df
```

```python
add_weekday_column(df8)
```
